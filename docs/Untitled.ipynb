```python
from spearmint.experiment import Experiment
from spearmint.hypothesis_test import HypothesisTest, CustomMetric
from spearmint.utils import generate_fake_observations
```

```python
test_data = generate_fake_observations(
    distribution="gaussian", n_treatments=6, n_observations=6 * 20, random_seed=123
)
exp = Experiment(test_data)
test_ab = HypothesisTest(
    metric='metric',
    control="A",
    variation="B",
    hypothesis="larger",
    inference_method="frequentist",
)
results_ab = exp.run_test(test_ab)
results_ab.display()
results_ab.visualize()
```
